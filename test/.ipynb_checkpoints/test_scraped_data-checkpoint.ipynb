```python
data = data.rename(columns={'RecordDate':'Record_TimeStamp'})
```

```python
data['Dept_Date'] = pd.to_datetime(data.Dept_Date, errors ='coerce')  # 'ignore would have kept invalid format string
```

```python
data['Record_Date'] = pd.to_datetime(data.Record_TimeStamp, errors ='coerce').dt.date
data['Record_Date'] = pd.to_datetime(data.Record_Date)
```

```python
data.groupby(['Dept_Date','Record_Date']).count()
```
